## 1. Peak-Wert bestimmen

```python
column = "C1_S5"

df = pd.read_csv("data/measurements.csv", sep=";", index_col="Temp./Â°C")

# Peakerkennung ---------------------------------------------------------

this_height = 0.3  # Grenze auf der y-Achse (0-1) 
this_prominence = 0.01 # [0.001 (mehr peaks) - 0.1 (weniger peaks)]

# Finden der Peaks
peaks, properties = find_peaks(df[column], height=this_height, prominence = this_prominence, width=1)


# Berechnung der ersten Ableitung der Daten -----------------------------
first_derivative = np.diff(df[column])

# Berechnung der zweiten Ableitung
second_derivative = np.gradient(first_derivative)

# Start- und Endpunkte bestimmen ----------------------------------------------------

# Initialisierung der Listen fÃ¼r die Startpunkte -------------------
start_points = []

# Iteriere durch jeden Peak, um den Startpunkt zu bestimmen
for peak_index in range(len(peaks)):
    peak = peaks[peak_index]
    # Suche rÃ¼ckwÃ¤rts nach dem letzten Punkt, bei dem der 1. Ableitung negativ ist
    start = None
    for i in range(peak, 0, -1):  # ab peak zurÃ¼ckwÃ¤rts (theoretisch bis Anfang)
        if first_derivative[i-1] < 0:
            start = i
            break
    if start is not None:
        start_points.append(start)
    else:
        start_points.append(0)


# Initialisierung der Listen fÃ¼r die Endpunkte ----------------------
end_points = []

# Iteriere durch jeden Peak, um den Endpunkt zu bestimmen
for peak_index in range(len(peaks)):
    peak = peaks[peak_index]
    # Suche vorwÃ¤rts nach dem ersten Punkt, bei dem der 1. Ableitung positiv wird
    end = None
    for i in range(peak, len(first_derivative)): # ab peak vorwÃ¤rts (theoretisch bis Ende)
        if first_derivative[i] > 0:
            end = i + 1  # BerÃ¼cksichtige np.diff Verschiebung
            break
    if end is not None:
        end_points.append(end)
    else:
        end_points.append(len(df) - 1)



# Startpunkte optimieren ----------------------------------------------
second_derivative_series = pd.Series(second_derivative)
second_derivative_series

# Initialisierung von improved_start_points 
improved_start_points = []

for i in range(len(start_points)):
    this_start = start_points[i]
    this_peak = peaks[i]

    # Eingrenzung der Series
    second_der_area = second_derivative_series.loc[range(this_start+1, this_peak)]

    # Umgekehrte For-Schleife durch die Series UM die negativen Werte "am Peak" zu lÃ¶schen
    for index in reversed(second_der_area.index):
        value = second_der_area[index]
        if value > 0:
            break;
        elif value <= 0:
            second_der_area = second_der_area.drop(index)

    new_start = this_start

    # Umgekehrte For-Schleife durch die Series, UM den Ãbergangs ins Negative zu finden (mit Toleranz) bzw. erste sehr nach an der 0
    for index in reversed(second_der_area.index):
        value = second_der_area[index]
        new_start = index
        
        if value <= 0.0001:
            if df[column].iloc[index] <= df[column].iloc[end_points[i]]:
                break
                
    improved_start_points.append(new_start)


# --------------------------------------------------------------------------------------
peaks_indices = df.index[peaks]

init_start_points_indices = df.index[start_points]
start_points_indices = df.index[improved_start_points]
end_points_indices = df.index[end_points]
```

```python
# Visualisierung ---------------------------------------------------------------------
plt.figure(figsize=(10, 6))
plt.plot(df[column], label="Daten")
plt.plot(peaks_indices, df[column].loc[peaks_indices], "x", label="Peak")

for i in range(len(start_points_indices)):
    plt.axvline(x=init_start_points_indices[i], color='pink', linestyle='--',
                label="init. Start" if init_start_points_indices[i] == init_start_points_indices[0] else "")
    plt.axvline(x=start_points_indices[i], color='r', linestyle='--',
                label="Start" if start_points_indices[i] == start_points_indices[0] else "")
    plt.axvline(x=end_points_indices[i], color='g', linestyle='--', 
                label="Ende" if end_points_indices[i] == end_points_indices[0] else "")

    plt.plot([start_points_indices[i], end_points_indices[i]], [df[column].loc[start_points_indices[i]], df[column].loc[end_points_indices[i]]], marker='o', label =f"Linie {i+1}")


plt.title(f"Peakerkennung fÃ¼r {column}")
plt.legend()
plt.show()
```

## 2. Peak Dataframe

```python
df_peak = pd.DataFrame(columns=["Series", "Peak_Temperature", "Start_Temperature", "End_Temperature", "Area"])
df_peak
```

```python
# Erstellen des Datasatzes
data = {"Series" : column,
        'Peak_Temperature': peaks_indices,
        'Start_Temperature': start_points_indices,
        'End_Temperature': end_points_indices,
        'Area': np.NaN}

if df_peak.empty:
    df_peak = pd.DataFrame(data)
else:
    df_peak = pd.concat([df_peak, pd.DataFrame(data)], ignore_index=True)


df_peak
```

## 3. Plot und FlÃ¤chenbrechnung

```python
# Plot der Daten
plt.figure(figsize=(10, 6))
plt.plot(df[column], label="Daten")

# Berechnung der FlÃ¤che zwischen den Peaks
areas = []
for i, row in df_peak[df_peak["Series"] == column].iterrows():
    start_index = df.index.get_loc(row['Start_Temperature'])
    end_index = df.index.get_loc(row['End_Temperature'])

    this_x = df.index[start_index:end_index]
    
    # Bereich unterhalb des Peaks
    area_below_peak = np.trapz(df[column].iloc[start_index:end_index], x=this_x)
    
    # Bereich zwischen der Linie von Start- und Endpunkt und dem Bereich unterhalb des Peaks
    line_values = np.linspace(df[column].iloc[start_index], df[column].iloc[end_index], end_index - start_index)
    area_below_line = np.trapz(line_values, x=this_x)
    
    # GesamtflÃ¤che
    total_area = area_below_peak - area_below_line
    areas.append(total_area)
    print(f"FlÃ¤che des Peaks {i+1}: {round(total_area, 5)}")
    
    # Einzeichnen der FlÃ¤che
    plt.fill_between(df.index[start_index:end_index], df[column].iloc[start_index:end_index], line_values, alpha=0.4, label=f"FlÃ¤che des Peaks {i+1}: {round(total_area, 4)}")

    plt.fill_between(df.index[start_index:end_index], df[column].iloc[start_index:end_index], alpha=0, label=f"Unterhalb des Peaks {i+1}: {round(area_below_peak, 4)}", hatch='\\')

    plt.fill_between(this_x, line_values, alpha=0.4, label=f"Unterhalb der Linie {i+1}: {round(area_below_line, 4)}")
    
plt.title(f"Peakerkennung fÃ¼r {column}")
plt.legend()
plt.show()
```
